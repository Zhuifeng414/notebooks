```python

import sqlite3
import pandas as pd

# This is a local copy of the webapp database that combines source, corpus, and model data
dbfile = '/Users/dylan/msds/capstone/battelle.db' 
db = sqlite3.connect(dbfile)



# We pull in the relevant table
DT = pd.read_sql_query('SELECT doc_id,topic_id,topic_weight FROM doctopic_long',db)
```

```python
DT.describe()
```

```python
# Use pandas pivot method to rearrange the data
DT = DT.pivot(index='doc_id', columns='topic_id', values='topic_weight')
```

```python
DT.iloc[0:10,]
```

```python
import numpy as np

# Take square root of every entry
DT_sqrt = DT.apply(np.sqrt)
```

```python
DT_sqrt.iloc[0:10]
```

```python
# Convert data frame to a numpy array
DT_sqrt_array = DT_sqrt.as_matrix()
```

```python
type(DT_sqrt_array)
```

```python
from sklearn import neighbors

# Build KD tree with Doc-Topic array (sqrt of all values taken)
DT_kdTree = neighbors.KDTree(DT_sqrt_array, leaf_size = 40)
```

```python
type(DT_kdTree)
```

```python
import pickle

file_Name = "DT_kdTree"
# open the file for writing
fileObject = open(file_Name,'wb') 

# this writes the KD_tree to file
pickle.dump(DT_kdTree,fileObject) 

# Close the file
fileObject.close
```
